# 桶排序

```python
# 变量无法统一 行数可以统一 修饰符不能统一 有自己的装饰器
# 变量先看数量，再看是否完全一致 不如直接取下来 不看变量名
# 所以应该一个函数对应一个 参数数量 参数list 行数 修饰符  注意参数list还是应该包括变量名 不然无法解决无类型名函数的问题
```

```python
#检查是否为JavaScript函数
count = 0
for i in range(len(snippet_function)):
    text = snippet_function.iloc[i]['code']
    if len(re.findall(r"let",text)) > 0:
        count += 1
count
```

```python
count = 0
for i in range(len(snippet_function)):
    text = snippet_function.iloc[i]['code']
    title = re.findall(r"function[\w|\s]*\([\w|\s|,]*\)[\w|\s]*{",text)[0].split('(')[1][:-1].split(")")
    para = title[0].replace('\n','')
    para = re.sub(' +', ' ', para).split(',')
    for item in para:
        if(len(item.strip().split()) == 1):
            count += 1
            break
    descriptor = title[1].replace('\n',' ').strip().split()
    print(para,end=' +')
    print(descriptor,end = ' ')
    print(count)
```

```python
snippet = []
code = []
bleu = []
snippet_code = list(snippet_function['dealed_code'])
code_code = list(code_function['dealed_code'])
snippet_code = snippet_code
for index,sni in enumerate(snippet_code):
    #print('match {}-----------------------------i'.format(index))
    big_bleu = []
    match_name = []    
    for j,onecode in enumerate(code_code):
        if abs(len(onecode)-len(sni)) < 20:
            temp = sentence_bleu([onecode],sni,smoothing_function=smooth.method1)
            if temp > 0.9:
                big_bleu.append(temp)
                match_name.append(j)                
    if len(big_bleu) > 0:
        snippet.append(index)
        bleu.append(big_bleu)
        code.append(match_name)
    print('match {}-----------------------------{}'.format(index,len(big_bleu)))
function = pd.DataFrame({
    'snippet':snippet,
    'bleu':bleu,
    'code':code
})
function.to_csv('function_match_v2.csv',index = False)    
```

```python
#匹配最近10个的匹配项，bleu和code分别用两个list存储
def takeSecond(elem):
    return elem[1]
snippet = []
code = []
bleu = []
snippet_code = list(snippet_function['dealed_code'])
code_code = list(code_function['dealed_code'])
snippet_code = snippet_code
for index,sni in enumerate(snippet_code[0:2]):#warming
    big_bleu = []
    match_name = []    
    temp_list = []#用于在list储存2值
    #遍历所有点，找到前十个节点，放入到中转list
    count = 10
    for j,onecode in enumerate(code_code):
        if abs(len(onecode)-len(sni)) < 20:
            temp = sentence_bleu([onecode],sni,smoothing_function=smooth.method1)
            temp_list.append([j,temp])
    temp_list.sort(key=takeSecond,reverse = True)#降序 根据相似度排序
    temp_list = temp_list[0:10]
    for item in temp_list:
        big_bleu.append(item[1])
        match_name.append(item[0])
    #if len(big_bleu) > 0:
    #已将10个储存好，放入最终list中
    snippet.append(index)
    bleu.append(big_bleu)
    code.append(match_name)
    print('match {}-----------------------------{}'.format(index,len(big_bleu)))
function = pd.DataFrame({
    'snippet':snippet,
    'bleu':bleu,
    'code':code
})
function.to_csv('function_match_10_item.csv',index = False) 
```
