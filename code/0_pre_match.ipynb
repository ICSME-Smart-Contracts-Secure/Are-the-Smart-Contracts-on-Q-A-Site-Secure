```python
def check_line(code):
    return len(code.splitlines()) > 1;
def classifier(text):
    function = re.search("function .* {",text)
    if function:
        contract = re.search("contract .* {",text)
        if contract:
            return 'contract'
        else:
            return 'function'
    else:
        return 'other'
def classifier2(text):
    function = re.search("function .* {",text)
    if function:
        contract = re.search("contract .*{",text)
        if contract:
            return 'contract'
        else:
            return 'function'
    else:
        return 'other'
```
